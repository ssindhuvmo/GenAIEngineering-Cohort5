### Pandas
pandas a python library that is extensively used for data analytics, ML and other Data fields  
It provides possibility to organise data into something called Dataframe, which acts as a table of data (2D)  
It is organised into rows and columns with named indices

```python
import pandas as pd
```

```python
# Create a data frame using the pandas function
DF = pd.DataFrame ({
                    'Name' : ['Vishnu', "Ram", 'Bharat'], 
                    'Id' : [234, 201, 78],
                    'Course' : ['ML', 'GenAI', "Python"],
                    'Batch' : ['C1', 'C2', 'C1']})
DF
```

```python
# Let us understand how is it organised
print ("Data type : \n", DF.dtypes)

print ("Index : \n", DF.index)

print ("Number of Records : \n", len(DF))

print ("Number of Fields : \n", len(DF.columns))
```

```python
# Access specific Column
Names = DF['Name']

print ("Type : \n", type (Names))
print ("Content : \n",  Names)
```

```python
# Access specific Row
Rec_1 = DF.iloc [1] # Access a row based on the index position

print ("Type : \n", type (Rec_1))
print ("Content : \n",  Rec_1)
```

```python
# Make specific Column in a data frame
Names = DF[['Name']]

print ("Type : \n", type (Names))
print ("Content : \n",  Names)
```

```python
# Filter out rows based on condition
Filtered = DF[DF['Batch'] == 'C1']
Filtered
```

```python
# Update an exising column values : just an assignment
DF['Id'] = [145, 156, 88]
DF
```

```python
# Add another column, again just like a variable assignment
DF ['Score'] = [78, 85, 80]
DF
```

```python
# Add Rows by concatenating
new_rec = pd.DataFrame ({
                        'Name' : ['Dileep', 'Phani'],
                        'Id' : [101, 102],
                        'Course' : ['GenAI', 'ML'],
                        'Batch' : ['C1', 'C2'],
                        'Score' : [80, 75]
})
new_rec
```

```python
DF = pd.concat ([DF, new_rec], ignore_index=True)
DF
```

```python
# Aggregating functions. See how the column is accessed
print ("Min Score : ", DF.Score.min())
print ("Max Score : ", DF.Score.max())
print ("Mean Score : ", DF.Score.mean())
print ("Median Score : ", DF.Score.median())
```

```python
# Check the Max score per course : Use Group by
grp = DF.groupby ('Course') ['Score'].max ().rename ("Max ")
grp
```

```python
# Read Data from CSV file
Rain_Data = pd.read_csv ('RainFall.csv')
print (Rain_Data.dtypes)
print (len(Rain_Data))
```

```python
# Filter the rows for annual rain fall higher than 1500
Rainy_Area = Rain_Data[Rain_Data['ANNUAL'] >= 1500]
Rainy_Area
```

```python
# Check Statistical summary of fields
print (Rain_Data[['ANNUAL', 'JAN', 'FEB', 'MAR']].describe())
```

```python
# Check Region Wise average
Summary = Rain_Data.groupby ('SUBDIVISION')['ANNUAL'].mean ().rename ("Annual Average").reset_index ()
Summary
```

```python
# Check min, max and average per region
summary = Rain_Data.groupby("SUBDIVISION")["ANNUAL"].agg(annual_min="min",annual_max="max",annual_avg="mean").reset_index()
summary
```

```python
# Filter the years where each sub division had annual rainfall above its average
filtered = Rain_Data[
    Rain_Data["ANNUAL"] >
    Rain_Data.groupby("SUBDIVISION")["ANNUAL"].transform("mean")
]
filtered
```

```python
# Make Quarter wise average
Rain_Data["Q1_avg"] = Rain_Data[["JAN", "FEB", "MAR"]].mean(axis=1)
Rain_Data["Q2_avg"] = Rain_Data[["APR", "MAY", "JUN"]].mean(axis=1)
Rain_Data["Q3_avg"] = Rain_Data[["JUL", "AUG", "SEP"]].mean(axis=1)
Rain_Data["Q4_avg"] = Rain_Data[["OCT", "NOV", "DEC"]].mean(axis=1)
Rain_Data
```
